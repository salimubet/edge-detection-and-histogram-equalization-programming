```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("image.jpg")
plt.imshow(img)
```

```python
imgGrey = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(imgGrey,cmap='gray')
```

```python
#medianlur = cv2.medianBlur(imgGrey,0)
#plt.imshow(medianlur,cmap='gray')
```

```python
edgesImg = np.zeros_like(imgGrey)
plt.imshow(edgesImg,cmap='gray')
```

```python
laplacian = cv2.Laplacian(edgesImg,cv2.CV_8U, -1)
plt.imshow(laplacian,cmap='gray')
```
